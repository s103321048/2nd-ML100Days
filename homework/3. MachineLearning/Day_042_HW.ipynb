```python
from sklearn import datasets, metrics
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()

train_x, test_x, train_y, test_y = train_test_split(iris.data, iris.target, test_size=0.25, random_state=4)

# my code
clf = DecisionTreeClassifier(criterion = 'entropy',
                             max_depth = None,
                             min_samples_split = 2,
                             min_samples_leaf = 1,
                            )

clf.fit(train_x, train_y)

y_pred = clf.predict(test_x)
```

### entropy

```python
#new
acc = metrics.accuracy_score(test_y, y_pred)
print("Accuarcy: ", acc)

print(iris.feature_names)
print("Feature Importance: ", clf.feature_importances_)
```

### gini

```python
acc = metrics.accuracy_score(test_y, y_pred)
print("Accuarcy: ", acc)

print(iris.feature_names)
print("Feature Importance: ", clf.feature_importances_)
```

```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
wine = datasets.load_wine()

train_x, test_x, train_y, test_y = train_test_split(wine.data, wine.target, test_size=0.25, random_state=4)

# my code
clf = DecisionTreeClassifier(criterion = 'gini',
                             max_depth = None,
                             min_samples_split = 3,
                             min_samples_leaf = 1,
                            )

clf.fit(train_x, train_y)

y_pred = clf.predict(test_x)
```

```python
acc = metrics.accuracy_score(test_y, y_pred)
print("Accuarcy: \n", acc)

print(wine.feature_names)
print("Feature Importance: \n", clf.feature_importances_)
```
