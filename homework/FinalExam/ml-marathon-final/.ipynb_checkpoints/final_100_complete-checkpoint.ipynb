```python
import numpy as np
import pandas as pd
out = pd.DataFrame(columns=['ID', 'Predicted'])

# 辨識每一張圖
for f in files:
    if f != '.DS_Store':
        img = image.load_img( DATASET_PATH+'/'+f, target_size=(224, 224))
        if img is None:
            continue
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis = 0)
        pred = net.predict(x)[0]
        top_inds = pred.argsort()[::-1][:5]
        #print(f[:3] + "  {:.8f}".format(pred[1]))
        print(f[:3], pred[1])
        
        
        name = int(f[:3])  #ID name
        out.loc[name] = f[:3], pred[1]  #save as dataframe
        
        
#        out.insert(str(f[:3]), str(pred[1]))
#        out["Predicted"].insert(lambda x:str(pred[1]))
```
